# Data init

```python
data = pd.read_csv(path, dtype="string")

data['Definition'] = data['Definition'].astype(str)
data['Word'] = data['Word'].astype(str)
```

## Preprocessing

*   Remove words with Underscores
*   Remove Double quotes
*   Capitalization: Convert all words to lowercase
*   Semicolons -> Multiple definitions

```python
data[~data['Word'].str.contains('_')].count()
```

```python
df = data[~data['Word'].str.contains('_')]
```

```python
df['Word'] = df['Word'].apply(lambda x: x.lower())
df['Definition'] = df['Definition'].apply(str.lower)
```

```python
df[~df['Word'].str.contains('\"')]
df = df.applymap(lambda x: x.replace('\"', ''))
df
```

```python
df[df['Definition'].str.contains(';')]
```

```python
df['Definition'] = df['Definition'].apply(lambda x : x.split(';'))
```

```python
df = df.explode('Definition')
```
